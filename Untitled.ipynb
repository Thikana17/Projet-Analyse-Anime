```python
import pandas as pd
# Chargement des données
df = pd.read_csv('animes (1).csv')
#Affiche les premières 
df.head()
```

```python
# Afficher le nombre de lignes et de colonnes 
df.shape
```

```python
# Afficher le nombre de valeur manquante par colonne
df.isnull().sum()
```

```python
# Remplacement des commentaires vides par "Inconnu" pour ne pas supprimer des lignes
df = df.fillna('Inconnu')
```

```python
# Afficher le nombre de valeur manquante par colonnes encore pour vérifier 
df.isnull().sum()
```

```python
# Supprime les doublons si il y en a 
df = df.drop_duplicates()
```

```python
# Affiche à nouveau le nombre de lignes et de colonnes 
df.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Indiquation de la taille du graphique 
plt.figure(figsize=(10, 6))
# On crée l'histogramme sur la colonne 'Note Globale'
sns.histplot(df['Note_Globale'], kde=True, color='skyblue')
plt.title('Distribution des Notes Globales des Animes')
plt.xlabel('Note sur 10')
plt.ylabel('Nombre d\'animes')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Indiquation de la taille du graphique 
plt.figure(figsize=(10, 6))
# On crée l'histogramme sur la colonne 'Note Globale'
sns.histplot(df['Note_Globale'], kde=True, color='skyblue')
plt.title('Distribution des Notes Globales des Animes')
plt.xlabel('Note sur 10')
plt.ylabel('Nombre d\'animes')
plt.show()
```

```python
plt.figure(figsize=(12, 7))
note_moyenne = df['Note_Globale'].mean()
sns.histplot(df['Note_Globale'], kde=True, color='#8e44ad', bins=15)
plt.axvline(note_moyenne, color='red', linestyle='--', linewidth=2, label=f'Moyenne : {note_moyenne:.2f}/10')
plt.title('Distribution des Notes : Où se situent la majorité des animes ?', fontsize=16, fontweight='bold')
plt.xlabel('Note sur 10', fontsize=12)
plt.ylabel("Nombre d'animes", fontsize=12)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# On compte combien il y a d'animes par type
df['type'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(8, 8))

# On ajoute un titre
plt.title("Répartition des animes par Type (TV, Film, OVA...)")
plt.ylabel('') 
# On cache le label moche sur le côté
plt.show()
```

```python
import matplotlib.pyplot as plt

# J'ai remplacé 'Type' par 'Source' 
df['Source'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(8, 8))

plt.title("Origine des animes (Manga, Original, etc.)")
plt.ylabel('')  # Pour faire propre
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))


sns.scatterplot(data=df, x='Nb_Episodes', y='Note_Globale')

plt.title("Lien entre la durée et la note")
plt.xlabel("Nombre d'épisodes")
plt.ylabel("Note Globale / 10")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# On compte les studios et on prend seulement les 10 premiers (head(10)) sinon le graphique serait illisible avec trop de barres
df['Studio'].value_counts().head(10).plot(kind='bar', color='orange')

plt.title("Top 10 des Studios les plus présents")
plt.ylabel("Nombre d'animes")
plt.xticks(rotation=45)
# On penche les noms pour qu'ils soient lisibles
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# On compare la note du meilleur épisode (en bas) avec celle du pire (à gauche)
sns.scatterplot(data=df, x='Note_Meilleur_Ep', y='Note_Pire_Ep')

# On ajoute une ligne diagonale pour aider à lire
# Si un point est sur la ligne, l'anime est parfaitement constant
plt.plot([0, 10], [0, 10], color='red', linestyle='--')

plt.title("Régularité : Écart entre le meilleur et le pire épisode")
plt.xlabel("Note du MEILLEUR épisode")
plt.ylabel("Note du PIRE épisode")
plt.show()
```

```python
# On sauvegarde les données nettoyées dans un nouveau fichier
# index=False permet de ne pas rajouter une colonne de numéros (0, 1, 2...) 
df.to_csv('animes_clean.csv', index=False)

print("Fichier sauvegardé avec succès !")
```

```python
# On ajoute deux options :
# sep=';'  -> Utilise des points-virgules pour qu'Excel comprenne les colonnes
# encoding='utf-8-sig' -> S'assure que les accents (é, à, è) s'affichent bien sur Windows

df.to_csv('animes_clean.csv', index=False, sep=';', encoding='utf-8-sig')

print("Fichier sauvegardé au format Excel Français !")
```
